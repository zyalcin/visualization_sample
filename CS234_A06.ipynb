## Task 3: Distance/Similarity Metrics

Write two functions, based on the formulas provided in class:


```python
import numpy as np
from math import sqrt
 
```

```python
def euclidean_distance(v1, v2):
    v1_array = np.array(v1) 
    v2_array = np.array(v2)
    return np.sqrt(np.sum(np.square(v1_array - v2_array)))
```

```python
euclidean_distance([3,4,2,1], [1,1,1,1])
```

```python
import math
```

```python
def cosine_similarity(v1,v2):
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)
```

```python
v1,v2 = [3, 45, 7, 2], [2, 54, 13, 15]
print(cosine_similarity(v1,v2))
```

## Task 4: Implement the kNN algorithm


`findNearestNeighbors` takes four parameters: one vector, one dataframe with instances (not including the vector that is the first parameter), the metric name (‘euclidean’ or ‘cosine’), and k (number of neighbors). After having calculated the distances between the given vector and every vector in the dataframe, it sorts them based on the meaning of the metric (for euclidean distance, we want the instances with the smallest distance; for the cosine similarity, we want the instances with the highest similarity). It finally returns the indices of the k nearest neighbors. For the purposes of this task, you should implement this function yourself and not use any library functions that do this work for you. In future assignments you can use library functions.


```python
def get_neighbors(v1, dfTrain, k, metric):
    metricDict = {'euclidean': euclidean_distance,
                  'cosine': cosine_similarity
                 }
    
    distancesLst = []
    for i, row in dfTrain.iterrows(): 
        f = metricDict[metric]        
        dist = f(v1, row)
        
        distancesLst.append((dist, i)) 
    
   
    reverse = True if metric=='cosine' else False
    distancesLst.sort(reverse=reverse)
    

    indices = [el[1] for el in distancesLst]
    return indices[:k] 
```

```python
v1 = dfTrain.loc[0]
dfTrain = dfTrain.drop(0)

get_neighbors(v1, dfTrain, 11, 'euclidean')
```

`predictLabel` takes two parameters: one dataframe that contains the labels,  and the list of indices returned by findNearestNeighbors. The function returns the predicted label based on the majority vote of the instances represented by their indices.


```python
def predictLabel(labelsDF, indices):
    neighbors = labelsDF[labelsDF.index.isin(indices)]
    return neighbors['sex'].mode()[0]
```
